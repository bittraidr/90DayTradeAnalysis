```python


df_trades2 = pd.read_csv('90.csv')
df_trades3 = pd.read_csv('90.csv')

# Assuming 'df' is your DataFrame
# Replace 'df' with the actual variable name if it's different
df = pd.read_csv('90.csv')

# List of columns to drop
columns_to_drop = ['User_Id', 'Order_Id', 'Transaction_Id', 'Primary_Asset', 
                   'Realized_Amount_For_Primary_Asset', 'Realized_Amount_For_Primary_Asset_In_USD_Value', 'Payment_Method', 'Withdrawal_Method', 'Additional_Note']

# Drop the specified columns
df_trades = df.drop(columns=columns_to_drop, axis=1)

# Now 'df' contains the DataFrame without the specified columns

df_trades.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with all the trades

# Plotting Buys
df_buy = df[df['Operation'] == 'Buy']
plt.scatter(df_buy['Time'], df_buy['Realized_Amount_For_Base_Asset'], color='green', label='Buys')

# Plotting Sells
df_sell = df[df['Operation'] == 'Sell']
plt.scatter(df_sell['Time'], df_sell['Realized_Amount_For_Base_Asset'], color='red', label='Sells')

# Formatting
plt.xlabel('Timestamp')
plt.ylabel('Realized Amount For Base Asset')
plt.title('Buys and Sells Over Time')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
